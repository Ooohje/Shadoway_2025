## 3) 교내 가로수 위치 표시

```python
file_path = "../artifacts/교내_가로수_정보.csv"
output_path = "../artifacts/campus_tree_map.html"

# csv 파일 읽기
df = pd.read_csv(file_path)
df.dropna(subset=['lat', 'lng'], inplace=True)

# 지도 생성
m = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=16)

# 좌표마다 마커
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=3,
        color='green',
        fill=True,
        popup=f"{row.get('memo', '')} - {row.get('ID', '')}"
    ).add_to(m)

m.save(output_path)
print("../artifacts/campus_tree_map.html에서 확인 가능")
```

## 5) 공간 데이터 + 속성 데이터 병합

```python
def load_and_merge_tree_data():
 
    tree_loc_path = "../artifacts/교내_가로수_정보.csv"
    tree_spec_path = "../artifacts/수종_정보.csv"

    trees_df = pd.read_csv(tree_loc_path)
    species_df = pd.read_csv(tree_spec_path)

    merged_df = pd.merge(trees_df, species_df, on='type', how='left')

    return merged_df

load_and_merge_tree_data()
```
